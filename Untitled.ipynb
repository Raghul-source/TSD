## Traffic Sign detection

### Load inputs from pickle

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline

#train_file = r'C:\Users\Raghul mani M\Downloads\train.p'
test_file = r'C:\Users\Raghul mani M\Downloads\archive\Test.csv'

#train_file = r'C:\Users\Raghul mani M\Desktop\train.p'
#test_file = r'C:\Users\Raghul mani M\Desktop\test.p'

with open('train.p', mode='rb') as f:
    train = pickle.load(f)
with open('test.p', mode='rb') as f:
    test = pickle.load(f)
    
X_train, y_train = train['features'], train['labels']
X_test, y_test = test['features'], test['labels']
print(len(X_train))
print(len(X_test))
num_images = 5
img_list = np.random.choice(len(y_train), num_images)
for i,idx in enumerate(img_list):
    plt.subplot(1, num_images,i+1)
    plt.imshow(X_train[idx,...])
    plt.title('Label: %d' % y_train[idx])
    
plt.tight_layout()
plt.show()
```

### Shuffle the data - X_train and y_train 


```python
def unison_shuffled_copies(X_train, y_train):
    assert len(X_train) == len(y_train)
    p = np.random.permutation(len(X_train))
    return X_train[p], y_train[p]
X_train,y_train=unison_shuffled_copies(X_train,y_train)
```

### Batch data and labels



```python
batch_size=5
def get_batch(features, labels, batch_size):
    """
    Split features and labels into batches
    """
    for start in range(0, len(features), batch_size):
        end = min(start + batch_size, len(features))
        yield features[start:end], labels[start:end]
    
```

```python
count = 0
for X_batch, y_batch in get_batch(X_train, y_train, batch_size):
    count += 1
    if count == 2:
        break
    
    
num_images = batch_size
for i in range(batch_size):
    plt.subplot(1, batch_size,i+1)
    plt.imshow(X_batch[i,...])
    plt.title('Label: %d' % y_batch[i])
    
```

```python
num_images = 5
img_list = np.random.choice(len(y_train), num_images)
for i in range(batch_size):
    plt.subplot(1, num_images,i+1)
    plt.imshow(X_train[i+5,...])
    plt.title('Label: %d' % y_train[i+5])
    
plt.tight_layout()
plt.show()
```

### Define a graph

```python
NUM_CLASSES = len(np.unique(y_train))
batch_size=64

tf.compat.v1.disable_eager_execution()
tf.compat.v1.reset_default_graph()
is_train = tf.compat.v1.placeholder(tf.bool)
_input = tf.compat.v1.placeholder(tf.float32, (None, 32, 32, 3))
_label = tf.compat.v1.placeholder(tf.int32, (None))
y = tf.one_hot(_label, NUM_CLASSES)
# Convolutional Layer #1 and Pooling Layer #1
net = tf.compat.v1.layers.conv2d(_input,filters=64,kernel_size=[3, 3],padding='same',activation=tf.nn.relu)
net = tf.compat.v1.layers.max_pooling2d(net,pool_size=[2, 2], strides=2)

# Convolutional Layer #2 and Pooling Layer #2
net = tf.compat.v1.layers.conv2d(inputs=net, filters=128, kernel_size=[3, 3], padding="same", activation=tf.nn.relu)
net = tf.compat.v1.layers.max_pooling2d(inputs=net, pool_size=[2, 2], strides=2)
net = tf.compat.v1.layers.dropout(net,rate=0.1,training=is_train)

# Convolutional Layer #3 and Pooling Layer #3
net = tf.compat.v1.layers.conv2d(inputs=net, filters=64, kernel_size=[3, 3], padding="same", activation=tf.nn.relu)
net = tf.compat.v1.layers.max_pooling2d(inputs=net, pool_size=[2, 2], strides=2)
net = tf.compat.v1.layers.dropout(net,rate=0.1,training=is_train)

print(net.get_shape().as_list())
flat = tf.reshape(net, [-1, 4 * 4 * 64])
#fully connected layer
dense = tf.compat.v1.layers.dense(inputs=flat, units=1024, activation=tf.nn.relu)
dense = tf.compat.v1.layers.dropout(dense,rate=0.1,training=is_train)
# Logits Layer
logits = tf.compat.v1.layers.dense(inputs=dense, units=NUM_CLASSES)

# Calculate loss
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y))

#optimizer(backpropagation)
#optimizer = tf.train.GradientDescentOptimizer(3e-4).minimize(loss)
# optimizer = tf.train.AdamOptimizer(3e-4).minimize(loss)
optimizer = tf.compat.v1.train.AdamOptimizer(3e-4).minimize(loss)

# Prediction (used during inference)
predictions = tf.argmax(logits, 1)

# Accuracy metric calculation
correct_prediction = tf.equal(predictions, tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

### RUN the session

```python
NUM_EPOCH = 25
with tf.compat.v1.Session() as sess:
    tf.compat.v1.global_variables_initializer().run()
    for epoch in range(NUM_EPOCH):
        count = 0
        for X_batch, y_batch in get_batch(X_train, y_train, batch_size):
            _, l, acc = sess.run([optimizer, loss,accuracy], {is_train:True,_input:X_batch, _label:y_batch})
            count += 1

            if count % 300==0:
                print ('Loss:{}, Accuracy:{}'.format(l, acc))
        s_acc=0.0
        for X_batch,y_batch in get_batch(X_test,y_test, 128):
            acc=sess.run(accuracy,{is_train:False,_input:X_batch,_label:y_batch})
            s_acc+=acc
            
        print('test accuracy:',(s_acc*128./len(y_test)))


print('done')
```
